```python
import pickle
import pandas as pd
import glob
```

```python
rep_files = glob.glob('./rep_result/*.pkl')
```

```python
rep_files
```

```python
with open(rep_files[0],'rb') as f:
    df = pickle.load(f)
    
for i in range(1,len(rep_files)):
    with open(rep_files[i],'rb') as f:
        tmp_df = pickle.load(f)
    df = pd.concat([df,tmp_df],ignore_index = True)
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(11,8)})
ax = sns.lineplot(data = df, 
                  x = 'Data Ratio',
                  y = "Prediction Pearson R", 
                  hue = 'Input Data',
                  errorbar='sd',
                  ci = None
                  #err_style="bars"
                  )
sns.move_legend(ax, "lower right")
```
