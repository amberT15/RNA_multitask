```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
celltype = 'HepG2'
```

```python
hepg2_df = pd.read_csv('./results/'+celltype+'_layer_model.csv')
hepg2_df['Model'] = 'Embedding CNN'
hepg2_df = hepg2_df.set_index('Layer')
hepg2_df["Mean Pearson's R"] = hepg2_df.mean(numeric_only=True, axis=1)
```

```python
hepg2_df['Mean Pearson\'s R'].max()
```

```python
fig, ax = plt.subplots(figsize=(6,6))
lineplot = sns.lineplot(data = hepg2_df , x='Layer', y = "Mean Pearson's R",hue = 'Model',ax = ax)
lineplot.set(title=celltype)

#onehot Residual
lineplot.axhline(0.4860,ls='--',c = 'k',)
plt.text(33,0.4860,'onehot Residual')

#onehot CNN
lineplot.axhline(0.3127,ls='--',c = 'g',)
plt.text(33,0.3127,'onehot CNN')

#MPRAnn
lineplot.axhline(0.301,ls='--',c = 'darkcyan',)
plt.text(33,0.295,'MPRAnn')

#zero shot
lineplot.axhline(0.027,ls='--',c = 'r',)
plt.text(33,0.027,'nucleotide transformer zero shot')

sns.move_legend(ax, "best", bbox_to_anchor=(1.4, 0))
plt.show()
```
