```python
import numpy as np
import h5py
import transformers
import gpn.mlm
from tqdm import tqdm
import torch
from scipy import special
import logomaker
import pandas as pd
import random
os.environ['CUDA_VISIBLE_DEVICES']='3'
```

```python
#dataset files
organisim = 'plants'
background = 'raw_sequence'
motif_set = np.loadtxt("/home/amber/multitask_RNA/data/motif/"+organisim+'/'+background+"/selected_motif.csv", delimiter="\n", dtype=str)
seq_data = h5py.File('/home/amber/multitask_RNA/data/motif/'+organisim+'/'+background+'/synthetic_seq.h5','r')
```

```python
#load models
model = transformers.AutoModelForMaskedLM.from_pretrained("gonzalobenegas/gpn-arabidopsis").to('cuda')
tokenizer = transformers.AutoTokenizer.from_pretrained("gonzalobenegas/gpn-arabidopsis")
mask_id = tokenizer.mask_token_id
model.eval();
```

```python
def meme_generate(W, output_file='meme.txt', prefix='filter'):
  """generate a meme file for a set of filters, W âˆˆ (N,L,A)"""

  # background frequency
  nt_freqs = [1./4 for i in range(4)]

  # open file for writing
  f = open(output_file, 'w')

  # print intro material
  f.write('MEME version 4\n')
  f.write('\n')
  f.write('ALPHABET= ACGT\n')
  f.write('\n')
  f.write('Background letter frequencies:\n')
  f.write('A %.4f C %.4f G %.4f T %.4f \n' % tuple(nt_freqs))
  f.write('\n')

  for j, pwm in enumerate(W):
    L, A = pwm.shape
    f.write('MOTIF %s%d \n' % (prefix, j))
    f.write('letter-probability matrix: alength= 4 w= %d nsites= %d \n' % (L, L))
    for i in range(L):
      f.write('%.4f %.4f %.4f %.4f \n' % tuple(pwm[i,:]))
    f.write('\n')

  f.close()
```

### Double checking synthesised sequence

```python
motif_list=[]
rc_list = []
for motif in tqdm(motif_set):
    seq = seq_data[motif]
    loci = seq_data[motif+'_loci']
    pred_motif = np.zeros((len(seq),4,len(loci[0])))
    for i in range(len(seq)):
        full_seq = seq[i].decode()
        full_token = tokenizer.encode(full_seq)
        mask_loci = loci[i]
        for c in range(len(mask_loci)):
            masked_token = full_token.copy()
            pred_motif[i,masked_token[mask_loci[c]]-2,c] = 1
    ori_motif = np.mean(pred_motif[:int(len(seq)/2)],axis=0)
    rc_motif = np.mean(pred_motif[int(len(seq)/2):],axis=0)
    motif_list.append(ori_motif)
    rc_list.append(rc_motif)
    break
pwm_df = pd.DataFrame(ori_motif.T,columns = ['A','C','G','T'])
logo_df=logomaker.transform_matrix(pwm_df,from_type='probability',to_type='information')
crp_logo = logomaker.Logo(logo_df,
                          shade_below=.5,
                          fade_below=.5,
                          font_name='Arial Rounded MT Bold')
```

### Run prediction for each loci in motif

```python
motif_list=[]
rc_list = []
for motif in tqdm(motif_set):
    seq = seq_data[motif]
    loci = seq_data[motif+'_loci']
    pred_motif = np.zeros((len(seq),len(loci[0]),4))
    for i in range(len(seq)):
        full_seq = seq[i].decode()
        full_token = tokenizer.encode(full_seq)
        mask_loci = loci[i]
        for j,c in enumerate(mask_loci):
            masked_token = full_token.copy()
            masked_token[c]=mask_id
            pred_loci = model(torch.tensor([masked_token]).to('cuda')).logits[0][c].cpu().detach().numpy()
            pred_motif[i,j,:] = special.softmax(pred_loci[2:],axis=0)
    ori_motif = np.mean(pred_motif[:int(len(seq)/2)],axis=0)
    rc_motif = np.mean(pred_motif[int(len(seq)/2):],axis=0)
    motif_list.append(ori_motif)
    rc_list.append(rc_motif)
```
