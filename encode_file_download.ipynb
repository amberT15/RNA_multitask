```python
import pandas as pd
import os
import subprocess
```

```python
K562_file = pd.read_csv('./data/eCLIP/K562/K562.tsv',sep='\t')
```

```python
K562_df = K562_file[K562_file['Biological replicate(s)'] == "1, 2"]
file_type_filter=(K562_df['Output type'] == 'peaks') & (K562_df['File assembly'] == 'GRCh38')& (K562_df['File format'] == 'bed narrowPeak')
K562_df = K562_df[file_type_filter]
```

```python
for idx,entry in K562_df.iterrows():
    link = entry['File download URL']
    cell_line = entry['Biosample term name']
    target = entry['Experiment target']
    cell_dir = './data/eCLIP/'+cell_line+'/'+target

    cmd = 'wget -O ' + os.path.join(cell_dir, target+'_replicate.bed.gz ') + link
    subprocess.call(cmd,shell='True')
    
cmd = 'gunzip ./data/eCLIP/*/*/*.bed.gz'
subprocess.call(cmd,shell ='True')
```
